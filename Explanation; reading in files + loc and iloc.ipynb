### Print head and then print the lines index with hours sleep >8 first 5, and in a nice and neat format

```python
# The Python way (Redundant; but handy for ubscure file types)
# By looping element-wise

c = []
with open(file) as f:
    for index, line in enumerate(f):
        Enough,Hours,PhoneReach,PhoneTime,Tired,Breakfast = line.strip().split(',')
        if index == 0:
            print(f'index    {Enough:10}{Hours:10}{PhoneReach:10}{PhoneTime:10}{Tired:10}{Breakfast:10}')
            
        else:
            if Hours != '':
                if int(Hours) > 8:
                    content = f'{index:10}{Enough:10}{Hours:10}{PhoneReach:10}{PhoneTime:10}{Tired:10}{Breakfast:10}'
                    c.append(content)

for i in range(5):
    print(c[i])
```
