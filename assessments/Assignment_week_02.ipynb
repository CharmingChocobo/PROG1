```python
from bokeh.io import output_notebook
from bokeh.plotting import figure, show
from bokeh.layouts import gridplot
from bokeh.io import output_notebook

from bokeh.plotting import ColumnDataSource

from bokeh.models import DatetimeTickFormatter, HoverTool
output_notebook()
```

```python
df.info()
```

```python
#CODE YOUR SOLUTION HERE
# We are going to ignore the ID column..

p = figure(title= 'Glucose over time', 
           x_axis_label= 'Time (unites?)', 
           x_axis_type= 'datetime',
           y_axis_label= 'Glucose (unites?)',
          )

p.title.text_font_size = "25px"

p.xaxis[0].formatter = DatetimeTickFormatter(minutes = ':%M')

p.line(x= df['time'],
       y= df['interpolated'],
       legend_label= 'Interpolated',
      )

p.line(x= df['time'],
      y= df['rolling'],
      color= 'red',
      line_color= 'pink',
      legend_label= 'Rolling mean',
      )

p.scatter(x= df['time'],
          y= df['glucose'],
          fill_color= 'orange',
          legend_label= 'Measured points'
         )

p.legend.location = 'top_left'

show(p)
```

```python
#CODE YOUR SOLUTION HERE
# Try np vs scipy interpolation methods

def Graph(df=df):
    # create ColumnDataSource based on dict
    source = ColumnDataSource(data=df)

    p = figure(title= 'Glucose over time', 
               x_axis_label= 'Time (unites?)', 
               x_axis_type= 'datetime',
               y_axis_label= 'Glucose (unites?)',
               tools= [HoverTool()],
               tooltips= 'Glucose at @x had value @y',
              )

    p.title.text_font_size = "25px"

    p.xaxis[0].formatter = DatetimeTickFormatter(minutes = ':%M')

    p.line(x= 'time',
           y= 'interpolated',
           legend_label= 'Interpolated',
           source=source,
          )

    p.line(x= 'time',
          y= 'rolling',
          color= 'red',
          line_color= 'pink',
          legend_label= 'Rolling mean',
          source=source,
          )

    p.scatter(x= 'time',
              y= 'glucose',
              fill_color= 'orange',
              legend_label= 'Measured points',
              source=source,
             )

    p.legend.location = 'top_left'

    return p
```

```python
show(Graph())
```
