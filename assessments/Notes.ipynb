```python
from bokeh.sampledata.iris import flowers
```

```python
df = flowers
```

```python
df.head()
```

```python
def RenderPlot(x= 'sepal_length', y= 'sepal_width'):
    p = figure()
    p.scatter(x= df[x], y= df[y])
    return p
    
```

```python
columns = df.columns[:-1]
columns
```

```python
inter_plot = pn.interact(RenderPlot, x=columns, y=columns)
inter_plot
```

```python
dashboard = pn.template.BootstrapTemplate(title= 'Braaap', sidebar_width= 300)
```

```python
dashboard.sidebar.append(inter_plot[0])
```

```python
dashboard.main.append(inter_plot[1])
```

```python
dashboard.show()
```
